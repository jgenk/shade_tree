```python
availability = pd.read_csv('availability.csv')
availability.loc[17,'School e-mail address:'] = 'courtney.m.edwards@vanderbilt.edu'
directory = pd.read_csv('directory.csv')
```

```python
availability.columns = ['timestamp','name','status','pap','email','comment','May','June','July','August']
months = ['May','June','July','August']
shifts = []
display(availability)
for col_name in months:
    col = availability[col_name]
    col = col.apply(lambda x: pd.Series(x.split(',')))
    shifts = shifts + col.loc[col.count(axis=1).sort_values().idxmax(),:].tolist()
```

```python
def check_available(row,shift):
    return ~row[months].str.contains(shift).any()

for shift in shifts:
    availability[shift] = availability.apply(lambda x: check_available(x,shift),axis=1)
```

```python
directory['in-charge'] = directory['Position Title'].isin(['Dispenser-in-charge','Co-Director of Dispensary'])
directory['med-runner'] = ((directory['Year'] != '2020') | directory['Email'].str.endswith('@uthsc.edu')) & (directory['Name'] != "Adam Stark")
scheduling_df = availability.merge(directory[['Email','in-charge','med-runner']],how='left',left_on='email',right_on='Email')

scheduling_df['pap'] = scheduling_df['pap'] == 'Yes'
```

```python
scheduling_df = scheduling_df.drop(['timestamp','status','email','comment'] + months,axis=1)
```

```python
roles = ['name','Email','pap','in-charge','med-runner']
days = [col for col in scheduling_df.columns.tolist() if col not in roles]
scheduling_df = scheduling_df[roles + days]
```

```python
scheduling_df.loc[:,days] = scheduling_df.loc[:,days].astype(bool)
availability_list = []
for day in days:
    available = scheduling_df[scheduling_df[day]][roles]
    print day
    display(available)
    incharge = available[available['in-charge']].name.tolist()
    paps = available[available['pap']].name.tolist()
    med_runner = available[(available['med-runner'])  & (~available['in-charge'])].name.tolist()
    other = available[~available[['in-charge','pap','med-runner']].any(axis=1)].name.tolist()
    series = pd.Series({'in-charge': ', '.join(incharge), 
                        'pap': ', '.join(paps),
                        'med-runner':', '.join(med_runner),
                        'other':', '.join(other)})
    series.name = day
    availability_list.append(series)

role_list = pd.DataFrame(availability_list)
```
