```python
import re
```

```python
def isPPline(line):
    return 'MCKESSON' in line

def extract_NDC(line):
    return re.search('[0-9]{11}',line).group(0)

def get_data(line):
    line = line.strip()
    description = re.search('^[A-Z\s\-]+',line).group(0)
    line = line[len(description):].strip()
    
    
    return [description] + line.split()
```
